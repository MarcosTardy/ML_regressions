##  Exercici 1
### Crea almenys dos models de regressió diferents per intentar predir el millor possible el preu de les vivendes (MEDV) de l'arxiu adjunt.

Attribute Information:

1. CRIM      per capita crime rate by town
2. ZN        proportion of residential land zoned for lots over 
             25,000 sq.ft.
3. INDUS     proportion of non-retail business acres per town
4. CHAS      Charles River dummy variable (= 1 if tract bounds 
             river; 0 otherwise)
5. NOX       nitric oxides concentration (parts per 10 million)
6. RM        average number of rooms per dwelling
7. AGE       proportion of owner-occupied units built prior to 1940
8. DIS       weighted distances to five Boston employment centres
9. RAD       index of accessibility to radial highways
10. TAX      full-value property-tax rate per $10,000

11. PTRATIO  pupil-teacher ratio by town
12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
             by town
13. LSTAT    % lower status of the population
14. MEDV     Median value of owner-occupied homes in $1000's

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
df = pd.read_csv("housing data (1).csv", sep=",",header=None)
df = df.sample(frac=1) # Desordena els registres del df
```

```python
df.columns=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
df 
#Assigno noms a les columnes segons les descripcions dels atributs.
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
correlacio = df.corr()
plt.figure(figsize=(12,6))
sns.heatmap(correlacio, cmap='coolwarm', annot=True)
plt.show()
```

El gràfic és una matriu de correlació modificada com un mapa de calor per visualitzar més fàcilment si dues variables tenen correlació o no. Quant més càlid sigui el color, més correlació existirà.

Com a exemple, si busquem la intercepció entre la variable "RAD" i la variable "TAX" es pot veure que el índex de correlació és de 0.91, per tant són proporcionals. En canvi, entre les variables "MEDV" i "LSTAT" hi ha una relació inversament proporcional.

La variables objectiu, és a dir la que vull predir, és la MEDV. Aquesta indica el valor mitg de les cases en milers de dólars. Per tant, el model de regressió que aplicaré té l'objectiu de predir, mitjançant la resta de variables, el valor mitjà de les cases. 

Si s'observa el gràfic d'adalt, és fàcil concluir que les úniques variables que tenen correlació amb la objectiu són RM (nombre d'habitacions per habitatge) i LSTAT (proporcio de persones de menor estatus de la població). Té sentit, perquè normalment quant més alt és el preu de la vivenda, més alt és el nombre d'habitacions que té i més baix és el percentatge de la població amb menor estatus.

Malgrat això, utilitzaré totes les variables ja que en el gràfic es mostra el coeficien de regressió lineal i no pas el de regressió polinòmica. Pot ser que hi hagi alguna variable que sí tingui relació amb la variable objectiu però que aquesta relació no sigui lineal. 

Abans de res he de fer la transformació de les dades aplicant RobusScaler i MinMaxScaler com vaig fer a l'sprint anterior. Vaig arribar a aquesta conclusió després d'estudir les funcions de densitat i els outliers de cada variable.

```python
from sklearn import preprocessing 
df[["CRIM", "ZN", "RM", "DIS", "B", "LSTAT", "PTRATIO"]]=preprocessing.RobustScaler().fit_transform(df[["CRIM", "ZN", "RM", "DIS", "B", "LSTAT", "PTRATIO"]])
```

```python
df[["INDUS","NOX","AGE","RAD","TAX"]]=preprocessing.MinMaxScaler().fit_transform(df[["INDUS","NOX","AGE","RAD","TAX"]])
```

Finalment convertiré la columna RAD en dummy. Es una variable categórica ja que classifica els registres segons índex d'accessibilitat a les autopistes segons que només pot ser 1,2,3 o 24.

```python
dummies = pd.get_dummies(df["RAD"], prefix="RAD")

# Agregamos las nuevas columnas dummy al DataFrame original
df = pd.concat([df, dummies], axis=1)
```

```python
df.head(5)
```

Un cop transformades les dades divideixo el dataset en train i test. Abans assigno les variables independents o "features" a la variable X i la dependent o objectiu a la variable y.

```python
from sklearn.datasets import make_blobs 
from sklearn.model_selection import train_test_split 
X = df.drop(["MEDV"], axis=1)
y= df[["MEDV"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape) 
```

Hi han 404 registres dins del training set i 102 registres dins del test set.

Començo fent un baseline per coneixer el valor màxim que podrà pendre l'error de la predicció. Un error més alt que el valor del average baseline error no serà acceptat.

```python
# Calcular la mitjana de la variable objectiu al conjunt d'entrenament.
baseline_value = y_train.mean()

# Crear un array camb el mateix tamany que el conjunt de prova y omplir-lo amb el val mitjà.
baseline_preds = np.full_like(y_test, baseline_value)

# Calcular el error de linia base 
baseline_errors = abs(baseline_preds - y_test.values) # abs per valor absolut.

# Mostrar el error mitjà de linia base
print('Average baseline error en milers de dòlars:', round(np.mean(baseline_errors), 3))
```

```python
print("El model, per ser acceptat, no podrà desviar-se en la seva predicció de més de", 
      round(np.mean(baseline_errors)*1000))
```

L'exercici 1 demana que apliqui dos models de regressió. Els models que he decidit fer són el Random forest i el de regressió lineal. El primer mètode de regressió que utilitzo és el RandomForest. Es pot aplicar ja que es tracta d'un problema de machine learning supervisat, ja que es tenen tant les variables independents com les dependents i, la variable objectiu és continua.

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor() 
rf.fit(X_train.values, y_train.values.ravel()) # Entrena el model.
#.ravel serveix per ajustar la forma de la llista.
```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
# Pull out one tree from the forest
tree = rf.estimators_[5]
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
# Pull out one tree from the forest
tree = rf.estimators_[5]
# Export the image to a dot file
export_graphviz(tree, out_file = 'tree.txt', 
                feature_names = list(X.columns), rounded = True, precision = 1)
# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('tree.txt')
# Write graph to a png file
graph.write_png('tree.png')
```

```python
predictions = rf.predict(X_test.values) # Fa 102 prediccion una per a cada valor de X_test.
errors=[]
for i in range(len(predictions)):
    errors.append(abs(predictions[i]-y_test.values[i]))
               # Compara el valor de cada predicció amb el real.
# Fer la mitjana de la llista d'errors.
print('Mean Absolute Error:', round(np.mean(errors)*1000), 'dolars.')
```

El error absolut mitjà és molt menor al límit calculat necessari per acceptar el model. 

```python
print("El resultat del model RandomForest sense aplicar CrossValidation és:",
      rf.score(X_test.values, y_test.values.ravel()))
```

Sembla que he creat un model de regressió molt bó, però aquesta mesura no representa correctament el bon funcionament del model. Per aconseguir una millor representació utilitzaré el cross validation. 

```python
from sklearn.model_selection import cross_val_score
# no vull fer CV sobre un RF ja entrenat, per això utilitzo una altra variable diferent a rf.
rf_noentr = RandomForestRegressor() 
# Faig el CV sobre un RF no entrenat
cv = cross_val_score(rf_noentr, X.values, y.values.ravel())
cv
```

El cross validation divideix el training set en 5. Cadascún d'aquests subconjunts conté 1/5 de les dades del X_train. Aquests subconjunts també es divideixen en 5 parts. D'aquesta manera cadascuna de les 5 parts és utilitzada una vegada per a evaluar el model.

A la primera iteració el cinqué subgrup és el utilitzat per a evaluar el model, en la segona és el quart el que s'utilitza per a evaluar-lo, i així succesivament. 

La llista d'adalt representa cadascuna d'aquestes 5 iteracions. 

```python
print("Un", round(cv.mean()*100,2), "% de les instancies han sigut encertades correctament.")
```

El segón model que aplicaré és el model basat en màquines de vectors de soport. L'objectiu d'aquest model és trobar la funció de regressió que minimitzi l'error entre les prediccions i els valor reals del conjunt de dades.

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train.values, y_train.values.ravel())
```

```python
print("El resultat del model SVR sense aplicar CrossValidation és:",
      svr.score(X_test.values, y_test.values.ravel())) 
```

No sembla un model massa fiable al costat del RandomForest. Aplico la validació creuada per conèixer amb major certesa la seva fiabilitat.

```python
from sklearn.model_selection import cross_val_score
svr_noentr = SVR()
cv = cross_val_score(svr_noentr, X.values, y.values.ravel())
cv
```

```python
print("El resultat del CrossValidation sobre el model SVR és:", cv.mean()) 
```

```python
predictions2 = svr.predict(X_test.values) # Fa 102 prediccion una per a cada valor de X_test.
# Per fer una gràfica comparant les prediccions del model amb la realitat.
```

```python
import matplotlib.pyplot as plt

# Crear una llista d'índexs per a l'eix x
indices = range(len(y_test.values))

# Crear una figura amb dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10,10))

# Grafica la primera gràfica al primer subplot
ax1.plot(indices, y_test.values, label='Valors reals')
ax1.plot(indices, predictions, label='Prediccions 1')
ax1.set_title('RandomForest', size=16)
ax1.set_ylabel('Valor')
ax1.legend()

# Grafica la segona gràfica al segón subplot
ax2.plot(indices, y_test.values, label='Valors reals')
ax2.plot(indices, predictions2, label='Prediccios 2')
ax2.set_title('SVR', size=16)
ax2.set_xlabel('Índex')
ax2.set_ylabel('Valor')
ax2.legend()

# Ajusta els espais entre subplots
plt.tight_layout()

# Mostrar la figura amb dos subplots
plt.show()


```

Es mostra d'una manera visual la exactitud del model. Es pot veure com per al model random forest, les linies practicament es solapen. Mentres que per al model SVR, la linia que representa les prediccions del model no acaba de concordar amb la linia que representa els valors reals.

## Exercici 3
### Entrena’ls utilitzant els diferents paràmetres que admeten per intentar millorar-ne la predicció.



Utilitzo grid search. El que fa aquest model es aplicar un cross validation per buscar la millor convinació de parametres possible per optimitzar el model, ja sigui el RandomForest o el SVR. El model es fa sobre un altre model NO entrenat. 

Per começar, observo els valor per defecte. A partir d'aquí buscaré valors propers. 

```python
rf.get_params()
```

```python
rff = RandomForestRegressor() 
#Torno a fer el Random Forest ja que no vull fer grid search amb el model ja entrenat.

from sklearn.model_selection import GridSearchCV
parametres = {"criterion":('poisson', 'squared_error', 'absolute_error', 'friedman_mse'),
             "n_estimators":(10,100,150)} 
              
gs = GridSearchCV(rff, parametres) #El scoring serà el que ve per defecte
gs.fit(X.values, y.values.ravel())
```

```python
gs.best_params_
```

```python
gs.best_estimator_
```

El millor criteri utilitzat per mesurar la qualitat d'una divisió a l'arbre de decisió és el poisson i el número d'estimadors més òptim és 150.


```python
gs.cv_results_
```

```python
#rf_gs = RandomForestRegressor(criterion='poisson', n_estimators=50) 
rff = RandomForestRegressor(criterion='poisson', n_estimators=150)
rff.fit(X_train, y_train.values.ravel()) # Entrena el model.
print("El resultat del model RF optimitzat sense aplicar CV és",
      rff.score(X_test.values, y_test.values.ravel()))
```

Comprovo que el resultat és millor, per tant és més òptim.
